# Playing Movies

This notebook lets you play a whole folder of Tiff files as a movie.

### User Inputs

```python
# Folder containing the images
images_folder = '/Users/priscilla/caiman_data/temp'
# /Users/priscilla/caiman_data/temp

first_file = None # None means start from the beginning (index 0)
last_file = None # None means go to the end (last index or index -1)
step = 3 # Every 'step' file will be taken (e.g., step=2 means take every second file)

downsample_ratio = 0.1  # Percentage of frames to keep
```

### Import Libraries

```python
import caiman as cm
from pathlib import Path
```

### Play Movie

```python
path = Path(images_folder)
fnames = sorted([p for p in path.rglob("[!.]?*.tif")])[first_file:last_file:step]

m_orig = cm.load_movie_chain(fnames)
m_orig.resize(1, 1, downsample_ratio).play(
    q_max=99.5,
    fr=30,
    magnification=2,
    plot_text=True,
    do_loop=False,
)  # play movie (press q to exit)
```

```python
path = Path(images_folder)
fnames = sorted([p for p in path.rglob("[!.]?*.mmap")])[first_file:last_file:step]

m_orig = cm.load_movie_chain(fnames)
m_orig.resize(1, 1, downsample_ratio).play(
    q_max=99.5,
    fr=30,
    magnification=2,
    plot_text=True,
    do_loop=False,
)  # play movie (press q to exit)
```
