# Playing Movies

This notebook lets you play the movies in the temp folder. 

(Make sure temp folder is empty before starting a new motion correction.)

### Import Libraries

```python
from utils import get_temp_movies
```

### User Inputs

```python
options = {
    "rigid": False,  # Show all temp files of the non-rigid motion correction
    "first_file": None,  # None means start from the beginning (index 0)
    "last_file": None,  # None means go to the end (last index or index -1)
    "step": 1,  # Every 'step' file will be taken (e.g., step=2 means take every second file)
    "downsample_ratio": 0.1,  # Percentage of frames to keep
    "batch_size": 3,  # Number of files to load at the same time
}
```

### Play Movie

```python
mcor_movie = get_temp_movies(**options)
```

```python
mcor_movie.play(
    q_max=99.5,
    fr=30,
    magnification=1,
    plot_text=True,
    do_loop=False,
)  # play movie (press q to exit)
```

Rerun the last cell to play the movie again.
